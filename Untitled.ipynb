### Multiclass Classification
If we have to deal with multiple class classification we have a neuron per class in a final layer<br>
We have to remember that NN accept only numbers so we might need to organize these classes<br>
i.e. we can use one-hot encoding (dummy variables)<br>

```python
ex_data = ['red', 'green', 'blue', 'red', 'red', 'green', 'red', 'blue' ,'green']
ex_df = pd.DataFrame(data = ex_data, columns=['Color'])
ex_df
```

Our NN is not gonna accept this data, so that's when we use one-hot encoding

```python
one_hot_df = pd.get_dummies(ex_df, dtype=int)
one_hot_df
```

Now it looks good :) <br>
We can also use it when dealing with non-exclusive classes <br>
When classes are exclusive we can use softmax activation function which sums all of the probabilities to 1 

### Cost Functions and Gradient Descent

We calculate the error just as usual, but our predicted value is the activation function output of the last layer.

We can express our error function as four arguments:  
C(W, B, Sr, Er), where:

- **W** is NN weights  
- **B** is biases  
- **Sr** is input of a single training sample  
- **Er** is desired output of the training sample  

But how do we figure out which weights lead us to the lowest cost? 

If we had a simple NN with only 1 weight **w** we'd like to minimize our loss function **C(w)**

If it was really a 1 dimensional function, we'd take the derivative and compare it to zero 

But in reality we're gonna deal with super complex n-dimensional functions and we can't calculate that

So we are to do a stochastic process and use gradient descent

We just calculate the slope of the function and move to the downwards direction of the slope by equal steps

Step size is known as a **learning rate** but we can adapt the step size known as **adaptive gradient descent**

## Preparing the data

```python
from sklearn.model_selection import train_test_split
```

Because of the way TF works we have to pass numpy arrays instead of pandas Series, that's why we add .values

```python
X = df[['feature1', 'feature2']].values
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Because we work with biases and weights we should normalize and scale data <br>
Apparently I dont have to scale the label, only the features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

Scaler.fit calculates parameters needed to perform scaling later on <br>

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# help(Sequential)
```

### Create a model

If something is densely connected it has a regular NN where every neuron is connected <br>
units arg in Dense means how many neurons in this layer <br>
activation chooses an activation function <br>
Now let's create a simple NN with 3 layers, notice that every arg called Dense is a single layer

```python
''' model = Sequential([Dense(units = 4, activation = 'relu'),
                   Dense(units = 2, activation = 'relu'),
                   Dense(units = 1)]) ''' 
```

There is another **better** way to do this: <br>
The last layer is built with only **1 neuron** because we try to predict a single value <br>
optimizer means an optimizing algorithm <br>
loss depends on what we try to predict (multi-class/binary classification or some value)

```python
model = Sequential()

model.add(Dense(units = 4, activation = 'relu'))
model.add(Dense(units = 4, activation = 'relu'))
model.add(Dense(units = 4, activation = 'relu'))

model.add(Dense(units = 1))

model.compile(optimizer = 'rmsprop', loss = 'mse')
```

### Fit the model
epochs means how many times do we go through the entire dataset <br>
We can see how MSE dynamically changes with each epoch

```python
model.fit(x = X_train, y = y_train, epochs = 250)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

Let's output the metric loss we decided to use (**MSE**) on test set

```python
model.evaluate(X_test, y_test, verbose = 0)
```

```python
model.evaluate(X_train, y_train, verbose = 0)
```

```python
test_predictions = model.predict(X_test)
type(test_predictions)
```

We need to make these predictions a pd Series to put it in the df

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
pred_df = pd.DataFrame(y_test, columns = ['Test True Y'])
pred_df.head()
```

```python
pred_df = pd.concat([pred_df, test_predictions], axis = 1)
pred_df.head()
```

Now we've got true and predicted values in one place

```python
pred_df.columns = ['Test True Y', 'Model Predictions']
pred_df.head()
```

So we can plot it out

```python
sns.scatterplot(x = 'Test True Y', y = 'Model Predictions', data = pred_df)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_true = pred_df['Test True Y'], y_pred = pred_df['Model Predictions'])
print(f'mean absolute error = {round(mae,4)}')
```

Now our goal is to determine if the score is desired or not <br>
To do so we compare our **mae** with original data

```python
df.describe()
```

```python
rate = (mae / df['price'].mean())*100
print(f'MAE rate is {rate}% of the mean value')
```

So we can say that our model makes good predictions

### Creating a model etc.

```python
X = filtered_df.drop(['price', 'yr_renovated', 'sqft_basement'], axis = 1).values
y = filtered_df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(units = 19, activation = 'relu'))
model.add(Dense(units = 19, activation = 'relu'))
model.add(Dense(units = 19, activation = 'relu'))
model.add(Dense(units = 19, activation = 'relu'))
model.add(Dense(units = 19, activation = 'relu'))


model.add(Dense(units = 1))

model.compile(optimizer = 'adam', loss = 'mse')
```

Now we're gonna add **validation data** which means in order to run the test data and check our loss after each epoch <br>
Of course this validation data is not gonna affect the weights and it's gonna be checked on test set

```python
model.fit(x = X_train, y = y_train,
          validation_data = (X_test, y_test), 
         batch_size = 128, epochs = 600)
```

We've got our model trained so let's find out what it learned <br>
To do so we can put **loss** and **validation loss** to the df and plot out how it behaves

```python
losses = pd.DataFrame(model.history.history)
```

This is the desired output when those two lines go together <br>
If the val_loss line had a spike (going up) it would mean that we overfitted our model

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```
